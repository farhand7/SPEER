```python
# import python packages
import pandas as pd
import numpy as np

# import src code
import ase_evaluation as ae
#import benchmark_posteriors as bp
import bootstrap as btstrp
import cross_validation as cv
import logistic_regression as lr
import naive_bayes as nb
import network as ntwk
import process as prcs
#import RIVER as rvr
import simulate_data as sim
from scipy import interp

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_data = []
for i in range(10):
    # generate simulated data
    s = sim.SimulateData("./test_output/", 'with_transfer', 0.4, 0.6, 0.01)
    s._run()
    
    # create a process object
    p = prcs.Process('./test_output', 0.1)
    p._process_simulated_data()
    
    n = ntwk.Network(p.train_list, p.test_list, p.tissues, p.genomic_features, 
                 with_transfer=True, output_dir="SPEER_output", 
                 lambda_hp_parent = None,
                 lambda_hp_children_dict = None,
                 e_distribution = 'cat')
    train_list, test_list, beta_parent, beta_children, phi = n.run()
    test_data.append(test_list)
    print(i)
```

```python
import sklearn
```

```python
len(test_data[0])
```

```python
fpr, tpr, auc, mean_tpr, mean_fpr, mean_auc = {}, {}, {}, {}, {}, {}
for i in range(len(test_data)):
    auc = 0
    fpr_local, tpr_local, auc_local = {}, {}, {}
    for j in range(len(test_data[i])):
        fpr_local[j], tpr_local[j], _ = sklearn.metrics.roc_curve(test_data[i][j]["z_labels"], test_data[i][j]["SPEER"])
        auc_local[j] = sklearn.metrics.roc_auc_score(test_data[i][j]["z_labels"], test_data[i][j]["SPEER"])
    mean_tpr[i] = 0.0
    mean_fpr[i] = np.linspace(0,1,100)
    for j in range(len(test_data[i])):
        mean_tpr[i] += interp(mean_fpr[i], fpr_local[j], tpr_local[j])
        mean_tpr[i][0] = 0.0
    mean_tpr[i] /= len(test_data[0])
    mean_tpr[i][-1] = 1.0
```

```python
auc = {}
for i in range(len(test_data)):
    auc[i] = sklearn.metrics.auc(mean_fpr[i], mean_tpr[i])
```

```python
%matplotlib inline
fig = plt.figure(figsize=(10,10))
#plt.style.use('seaborn-talk')
for i in range(len(test_data)):
    ax = plt.plot(mean_fpr[i], mean_tpr[i], label=' {0:0.3f}'
              ''.format(auc[i]), linewidth=3, c='blue')
ax = plt.plot([0,1], [0,1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=20)
sns.set_context("paper")
sns.set_palette("deep")
sns.set(font='serif')
sns.set_style("white", {"font.family": "serif", "font.serif": ["Times", "Palatino", "serif"]})
#ax.spines['top'].set_visible(False)
#ax.spines['right'].set_visible(False)
plt.ylabel('True Positive Rate', fontsize=20)
#plt.title('Receiver operating characteristic')
plt.legend(loc="lower right", frameon=False, prop={'size':16})
plt.tight_layout()
```
