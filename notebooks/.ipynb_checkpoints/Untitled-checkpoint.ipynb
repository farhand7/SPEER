```python
%matplotlib inline
import matplotlib.pyplot as plt
fpr, tpr, auc = dict(), dict(), dict()
# ROC for TL
#inds = pd.notnull(test_all).all(1).nonzero()[0]
#test_all = test_all.ix[inds]
#test_all = test_all.ix[inds]

fpr[0], tpr[0], _ = sklearn.metrics.roc_curve(x["N2_expr_label"], x["posterior"])
auc[0] = sklearn.metrics.auc(fpr[0], tpr[0])

# ROC for RIVER
fpr[1], tpr[1], _ = sklearn.metrics.roc_curve(x["N2_expr_label"], x["posterior_RIVER"])
auc[1] = sklearn.metrics.auc(fpr[1], tpr[1])

# ROC for tissue-specific genome only model
fpr[2], tpr[2], _ = sklearn.metrics.roc_curve(x["N2_expr_label"], x["posterior_genome_only"])
auc[2] = sklearn.metrics.auc(fpr[2], tpr[2])

# ROC for shared tissue genome only model
fpr[3], tpr[3], _ = sklearn.metrics.roc_curve(x["N2_expr_label"], x["posterior_genome_only_shared"])
auc[3] = sklearn.metrics.auc(fpr[3], tpr[3])

# ROC for N1 = N2 classifier
fpr[4], tpr[4], _ = sklearn.metrics.roc_curve(x["N2_expr_label"], x["expr_label"])
auc[4] = sklearn.metrics.auc(fpr[4], tpr[4])

# ROC for N1 = N2 classifier
fpr[5], tpr[5], _ = sklearn.metrics.roc_curve(x["N2_expr_label"], x["posterior_avg"])
auc[5] = sklearn.metrics.auc(fpr[5], tpr[5])



plt.figure()
plt.style.use('seaborn-talk')
plt.plot(fpr[0], tpr[0], label='Multi-task = {0:0.3f}'
         ''.format(auc[0]), linewidth=2)
plt.plot(fpr[1], tpr[1], label='RIVER = {0:0.3f}'
         ''.format(auc[1]), linewidth=2)
plt.plot(fpr[2], tpr[2], label='Tissue-spec. LR = {0:0.3f}'
         ''.format(auc[2]), linewidth=2)
plt.plot(fpr[3], tpr[3], label='Shared LR = {0:0.3f}'
         ''.format(auc[3]), linewidth=2)
plt.plot(fpr[4], tpr[4], label='N1=N2 = {0:0.3f}'
         ''.format(auc[4]), linewidth=2)
plt.plot(fpr[5], tpr[5], label='average rate = {0:0.3f}'
         ''.format(auc[5]), linewidth=2)

plt.plot([0,1], [0,1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
#plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
```
