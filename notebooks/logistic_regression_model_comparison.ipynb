This files compares logistic regression for the tissue-specific vs shared tissue model.

The labels are discretized expression values using a median z-score threshold 1.5.

We will compare the accuracy between four tissues and a median of the 4 tissues. Each model will use the same 
samples.

```python
# discretize data
#Y_cont = np.abs(train_df[['Adipose_Subcutaneous', 'Adipose_Visceral_Omentum', 'Adrenal_Gland', 'Esophagus_Muscularis', 'median']])
g_start = list(artery_df.columns).index("median") + 1
Y_cont = np.abs(artery_df["median"]).reshape(-1,1)
Y = sklearn.preprocessing.binarize(Y_cont, threshold=2.5)
G = sklearn.preprocessing.scale(artery_df.iloc[:,g_start:], axis=0)

print Y.shape, G.shape
```

```python
G_train, G_test, y_train, y_test = sklearn.cross_validation.train_test_split(G, Y[:,0], test_size=0.25, random_state=42)
l = sklearn.linear_model.LogisticRegression(penalty='l2', tol=.0001, fit_intercept=True)
l.fit(G_train, y_train)
sklearn.metrics.roc_auc_score(y_test, l.predict_proba(G_test)[:,1])
```
