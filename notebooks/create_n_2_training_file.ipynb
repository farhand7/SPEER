This file generates the N=2 Training file.

The training file has the following tab delimited format:

s1 g1 s2 g2 t1(expr1,expr2) t2(expr1,expr2) ... tn(expr1,expr2) median(expr1,expr2)

GTEX-P4PQ	ENSG00000100038.15  GTEX-ZXES	ENSG00000100038.15 2.6,3.1 2.3,2.8 2.9,2.1 2.1,2.3

```python
import numpy as np
import pandas as pd
import sys
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn import preprocessing
import seaborn as sns; sns.set(style="white", color_codes=True)
import scipy
```

```python
df = pd.read_table('../../../../../../data/train/v6/44tissues_10ktss_median.w_pc.g.v6p.txt', index_col=(1,0))
del df["Unnamed: 122"]
```

```python
df
```

```python
# identify all N>=2 pairs
df["gene"] = df.index.get_level_values('gene')
#df['is_duplicated'] = df.groupby('gene').duplicated(df.columns)

df['is_duplicated'] = df.duplicated(df.columns)
x = df[df['is_duplicated']].groupby(['gene', 'max_GerpS_10kb','max_verPhCons_10kb','max_GC_10kb', 'max_CpG_10kb', 'max_priPhCons_10kb' ]).size()
#x = df[df['is_duplicated']].groupby([df.columns[:-1]]).size()

x +=1 # original + duplicate
```

```python
x
```

```python
#df_swapped = df.swaplevel('indiv', 'gene')
df_frame = pd.DataFrame()
count = 0
for ind in x.index:
    gene = str(ind[0])
    max_gerp = ind[1]
    max_verphcons = ind[2]
    max_gc = ind[3]
    max_cpg = ind[4]
    max_priphcons = ind[5]
    num_variants = x[ind]
    y = df.ix[gene]
    
    subjects = y[y['max_GerpS_10kb' and 'max_verPhCons_10kb' and 'max_GC_10kb' and 'max_CpG_10kb' and
                  'max_priPhCons_10kb'] == (max_gerp and max_verphcons and max_gc and max_cpg and max_priphcons)].index.tolist()
    
    #print (y['max_GerpS_10kb' and 'max_verPhCons_10kb'] == max_gerp and y['max_verPhCons_10kb'] == max_verphcons).all()
    #subjects = y[y['max_GerpS_10kb'] == max_gerp].index.tolist()
    subjects.insert(0,gene)
    df_frame = df_frame.append(pd.Series(subjects), ignore_index=True)
```
